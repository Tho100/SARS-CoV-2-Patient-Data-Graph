```python
import pandas as pd 
import matplotlib.pyplot as plt 

plt.style.use('ggplot')

data = pd.read_excel("C:\\users\\HP\\Downloads\\covid_case.xlsx")

positive = (data[data['Positive SARS-CoV-2'] == True]['Patient Name']).head(7)
positive_bdate = (data[data['Positive SARS-CoV-2'] == True]['Birthdate']).head(7)

plt.bar(positive,positive_bdate)
plt.title('Positive SARS-CoV-2')
plt.xlabel('Patient Name')
plt.ylabel('Patient Birthdate')
plt.show()

false_positive = (data[data['Positive SARS-CoV-2'] == False]['Patient Name']).head(7)
false_positive_bdate = (data[data['Positive SARS-CoV-2'] == False]['Birthdate']).head(7)

plt.bar(false_positive,false_positive_bdate,color=['blue'])
plt.title('False Positive Positive SARS-CoV-2')
plt.xlabel('Patient Name')
plt.ylabel('Patient Birthdate')
plt.show()

positive_occupation = (data[data['Positive SARS-CoV-2'] == True]['Occupation']).head(7)

plt.bar(positive,positive_occupation)
plt.title('Positive SARS-CoV-2 Occupation')
plt.xlabel('Patient Name')
plt.ylabel('Occupation')
plt.show()

negative_occupation = (data[data['Positive SARS-CoV-2'] == False]['Occupation']).head(7)

plt.bar(false_positive,negative_occupation,color=['blue'])
plt.title('False Positive SARS-CoV-2 Occupation')
plt.xlabel('Patient Name')
plt.ylabel('Occupation')
plt.show()
```
